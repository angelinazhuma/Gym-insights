# Model Training

### Importing Libraries and Loading Dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("gym_members_exercise_cleaned.csv")
```

### Data Preprocessing: Encoding, Feature Selection, and Standardization

```python
# Encode categorical variables
data['Gender'] = LabelEncoder().fit_transform(data['Gender'])
data['Workout Type'] = LabelEncoder().fit_transform(data['Workout Type'])

# Features and targets
features = data.drop(columns=['Experience Level', 'Calories Burned'])
classification_target = data['Experience Level']
regression_target = data['Calories Burned']

# Standardize numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

###  Data Splitting for Classification and Regression

```python
# Split data for classification
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    scaled_features, classification_target, test_size=0.2, random_state=42)

# Split data for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    scaled_features, regression_target, test_size=0.2, random_state=42)
```

### Train Classification Model

```python
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train_cls, y_train_cls)
y_pred_cls = classifier.predict(X_test_cls)
classification_accuracy = accuracy_score(y_test_cls, y_pred_cls)
```

### Feature Importance for Classification

```python
feature_importances_cls = classifier.feature_importances_
```

### Train Regression Model

```python
regressor = RandomForestRegressor(random_state=42)
regressor.fit(X_train_reg, y_train_reg)
y_pred_reg = regressor.predict(X_test_reg)
regression_rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
```

```python
# Feature Importance for Regression
feature_importances_reg = regressor.feature_importances_
```

```python
# Print Results
print("Classification Model Accuracy: {:.2f}%".format(classification_accuracy * 100))
print("Regression Model RMSE: {:.2f}".format(regression_rmse))
```

### Visualization of Feature Importances

```python
import matplotlib.pyplot as plt

# Plot Predicted vs Actual for regression model
plt.figure(figsize=(10, 6))
plt.scatter(y_test_reg, y_pred_reg, alpha=0.6, edgecolor='k')
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], 'r--', lw=2)
plt.xlabel('Actual Calories Burned')
plt.ylabel('Predicted Calories Burned')
plt.title('Predicted vs. Actual - Regression Model')
plt.show()
```

```python
features_names = features.columns

plt.figure(figsize=(12, 6))

# Classification Feature Importances
plt.subplot(1, 2, 1)
sns.barplot(x=feature_importances_cls, y=features_names)
plt.title('Feature Importances - Classification')
plt.xlabel('Importance')
plt.ylabel('Feature')

# Regression Feature Importances
plt.subplot(1, 2, 2)
sns.barplot(x=feature_importances_reg, y=features_names)
plt.title('Feature Importances - Regression')
plt.xlabel('Importance')
plt.ylabel('Feature')

plt.tight_layout()
plt.show()
```

```python
# Calculate residuals (errors between predicted and actual values)
residuals = y_test_reg - y_pred_reg

# Plot residuals
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_reg, residuals, alpha=0.6, edgecolor='k')
plt.axhline(0, color='red', linestyle='--', lw=2)  # Line where residuals = 0
plt.xlabel('Predicted Calories Burned')
plt.ylabel('Residuals')
plt.title('Residual Plot - Regression Model')
plt.show()
```

```python
# Distribution of Classification Target (Experience Level)
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Experience Level')
plt.title("Distribution of Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Count")
plt.show()

# Distribution of Regression Target (Calories Burned)
plt.figure(figsize=(10, 6))
sns.histplot(data['Calories Burned'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Calories Burned")
plt.xlabel("Calories Burned")
plt.ylabel("Frequency")
plt.show()
```

```python
# Correlation Heatmap for Numerical Features
plt.figure(figsize=(12, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

```python
importance_df = pd.DataFrame({
    'Feature': features.columns,
    'Classification Importance': feature_importances_cls,
    'Regression Importance': feature_importances_reg
}).set_index('Feature')

# Normalize importance values for better comparison
importance_df = importance_df / importance_df.max()

plt.figure(figsize=(12, 8))
sns.heatmap(importance_df, annot=True, cmap='coolwarm', cbar=True)
plt.title('Feature Importance Heatmap - Classification vs Regression')
plt.show()
```

### Hyperparameter Tuning for Classification and Regression Models

```python
# Hyperparameter Tuning - Classification
param_grid_cls = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_search_cls = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_cls, cv=3)
grid_search_cls.fit(X_train_cls, y_train_cls)
best_classifier = grid_search_cls.best_estimator_
y_pred_cls_tuned = best_classifier.predict(X_test_cls)
classification_accuracy_tuned = accuracy_score(y_test_cls, y_pred_cls_tuned)

# Hyperparameter Tuning - Regression
param_grid_reg = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_search_reg = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_reg, cv=3)
grid_search_reg.fit(X_train_reg, y_train_reg)
best_regressor = grid_search_reg.best_estimator_
y_pred_reg_tuned = best_regressor.predict(X_test_reg)
regression_rmse_tuned = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg_tuned))

# Print Tuned Results
print("Tuned Classification Model Accuracy: {:.2f}%".format(classification_accuracy_tuned * 100))
print("Tuned Regression Model RMSE: {:.2f}".format(regression_rmse_tuned))
```

### Conclusion: Classification and Regression Models for Gym Member Data

**Classification Model: Predicting Experience Level**

- **Input Data**: The classification model takes features such as gender, workout type, age, weight, height, and BMI to predict the Experience Level of gym members.
- **Training**: The model is trained using a Random Forest Classifier.
- **Prediction**: After training, the model classifies each gym members experience level based on their workout and demographic data. The classification output is a label for the experience level (0 for Beginner, 1 for Intermediate, 2 for Expert).
- **Decision Making**: The Random Forest classifier builds multiple decision trees, each making a prediction.
- **Accuracy**: The tuned classification model achieved an impressive 90.62% accuracy in predicting the gym members' experience levels.
- **Key Insights**:
    - The classification model effectively categorizes gym members into distinct experience levels based on their features.
    - The Random Forest classifier captures complex relationships between features and experience level.
    - Hyperparameter tuning further improved the model's performance, enhancing its classification accuracy.
**Example**:  
For a gym member with the following features:
- Gender: Male  
- Workout Type: Strength Training  
- Age: 25  
- BMI: 24  
- Weight: 75 kg  
The model will predict the experience level based on these factors. If the majority of trees predict an Intermediate level, the model outputs 1.
**Regression Model: Predicting Calories Burned**
- **Input Data**: The regression model uses features like age, weight, height, workout type, and session duration to predict the number of Calories Burned during a workout session.
- **Training**: The model is trained using a Random Forest Regressor, which is well-suited for predicting continuous variables like calories burned. It works by averaging predictions from multiple decision trees to minimize prediction errors.
- **Prediction**: After training, the model predicts the number of calories burned for each test case based on the input features.
- **Decision Making**: The Random Forest regressor uses the aggregated predictions from all decision trees to output the final calorie count. Each tree predicts a value, and the overall result is the average of all predictions.
- **RMSE**: The tuned regression model achieved a Root Mean Squared Error of 39.11, which reflects the model's accuracy in predicting the number of calories burned during workout sessions. The lower the RMSE, the better the models predictive performance.
- **Key Insights**:
    - The regression model predicts calorie expenditure accurately, helping users estimate their energy expenditure during workouts.
    - Key features influencing calorie burn include workout type, session duration, and body composition.
    - Hyperparameter tuning further improved the accuracy of the regression model by optimizing the number of trees and their depth.
**Example**:  
For a gym member with the following features:
- Gender: Female  
- Weight: 60 kg  
- Session Duration: 1 hour  
- Workout Type: Cardio  
The model will predict the number of calories burned based on these factors. If the model predicts 350 calories, the output is a continuous value representing the estimated calories burned during that workout.

- **Classification Model Accuracy**: The Random Forest Classifier achieved a strong accuracy in predicting experience levels, and tuning the model further improved its performance.
  
- **Regression Model RMSE**: The Random Forest Regressor showed low root mean squared error (RMSE), meaning it makes accurate predictions for calories burned during workouts.

- **Hyperparameter Tuning**: Both models benefited from hyperparameter tuning, resulting in improved prediction performance. For the classification model, accuracy improved, while for the regression model, RMSE decreased significantly.

By using Random Forest models for classification and regression tasks, we were able to make robust predictions regarding gym member experience levels and calories burned.
